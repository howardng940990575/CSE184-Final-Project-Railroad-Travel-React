```python
import pandas as pd
```

```python
shanghai_station = pd.read_csv("shanghai_train_stations.csv")
zhejiang_station = pd.read_csv("zhejiang_train_stations.csv")
jiangsu_station = pd.read_csv("jiangsu_train_stations.csv")

shanghai_english = pd.read_csv("shanghai_english.csv")
zhejiang_english = pd.read_csv("zhejiang_english.csv")
jiangsu_english = pd.read_csv("jiangsu_english.csv")
```

```python
result_shanghai = pd.concat([shanghai_station, shanghai_english], axis=1)
result_zhejiang = pd.concat([zhejiang_station, zhejiang_english], axis=1)
result_jiangsu = pd.concat([jiangsu_station, jiangsu_english], axis=1)
```

```python
result_jiangsu
```

```python
result_shanghai.to_json("shanghai_station_english.json")
result_jiangsu.to_json("jiangsu_station_english.json")
result_zhejiang.to_json("zhejiang_station_english.json")
```
